#3. Masa invariante

Es un invariante de Lorentz y se obtiene en fsica de colisionadores a travs de

$$M^2=2p_{T_1}p_{T2}[cosh(\eta_2-\eta_1)-cos(\phi_1-\phi_2)] $$



```python
def mtransversa(pt1,pt2,eta1,eta2,phi1,phi2):
  return np.sqrt(2*pt1*pt2*(np.cosh(eta2-eta1)-np.cos(phi1-phi2)))

mtransv=mtransversa(eventos['pt1'],eventos['pt2'],eventos['eta1'],eventos['eta2'],eventos['phi1'],eventos['phi2'])

fig_5=px.histogram(mtransv,mtransv)
fig_5.update_layout(title=r"$\text{Masa invariante} \ M, \ \text{%d bins}$" %1000,
                    xaxis_title=r"$GeV/c$",yaxis_title="Eventos")
fig_5.show()
```

#6. Distribucin de $\Delta \phi $

```python
def deltaphi(phi1,phi2):
  return np.sqrt((phi1-phi2)*(phi1-phi2))

dphi=deltaphi(eventos['phi1'],eventos['phi2'])

dphiD=px.histogram(dphi,dphi,nbins=1000)
dphiD.update_layout(title=r"$\Delta \phi \ \text{Distribution} , \ \text{%d bins}$" %1000,
                    xaxis_title=r"$\Delta \phi \ \text{(en Radianes)}$",yaxis_title="Nm. de Eventos")
dphiD.show()

```
