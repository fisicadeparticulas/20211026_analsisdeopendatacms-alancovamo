#3. Masa invariante

Es un invariante de Lorentz y se obtiene en física de colisionadores a través de

$$M^2=2p_{T_1}p_{T2}[cosh(\eta_2-\eta_1)-cos(\phi_1-\phi_2)] $$



```python
def mtransversa(pt1,pt2,eta1,eta2,phi1,phi2):
  return np.sqrt(2*pt1*pt2*(np.cosh(eta2-eta1)-np.cos(phi1-phi2)))

mtransv=mtransversa(eventos['pt1'],eventos['pt2'],eventos['eta1'],eventos['eta2'],eventos['phi1'],eventos['phi2'])

fig_5=px.histogram(mtransv,mtransv)
fig_5.update_layout(title=r"$\text{Masa invariante} \ M, \ \text{%d bins}$" %1000,
                    xaxis_title=r"$GeV/c$",yaxis_title="Eventos")
fig_5.show()
```

# Se decide hacer un corte para iso$_{1,2}<\varepsilon$

```python

epsilon=1
eventos_filtrados=eventos[abs(eventos.iso1) < epsilon]
eventos_filtrados=eventos_filtrados[abs(eventos_filtrados.iso2) < epsilon]
eventos_filtrados.shape[0]

print('%d eventos cumplen los criterios' %eventos_filtrados.shape[0])
```

```python
mtransv_limpiada=mtransversa(eventos_filtrados['pt1'],eventos_filtrados['pt2'],eventos_filtrados['eta1'],
                             eventos_filtrados['eta2'],eventos_filtrados['phi1'],eventos_filtrados['phi2'])

fig_l=px.histogram(mtransv_limpiada,mtransv_limpiada)
fig_l.update_layout(title=r"$\text{Masa invariante (limpiada)} \ M, \ \text{%d bins}$" %1000,
                    xaxis_title=r"$GeV/c$",yaxis_title="Eventos")
fig_l.show()
fig_5.show()
```
